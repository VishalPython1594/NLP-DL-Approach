# Step 5 : Apply data prepreparation on training data

```python
import string
seg = r'https:\S+'
lemma = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))
```

```python
def clean(doc):

    doc = re.sub(seg, '', doc)
    doc = doc.replace('</br>', ' ')
    doc = re.sub(r'[^\x00-\x7F]+', '', doc)
    doc = doc.lower()
    doc = ''.join([char for char in doc if char not in string.punctuation and not char.isdigit()])
    doc_token = nltk.word_tokenize(doc)
    lemma_doc = [lemma.lemmatize(token) for token in doc_token]
    stop_doc = [word for word in lemma_doc if word.lower() not in stop_words]
    return ' '.join(stop_doc)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vect = TfidfVectorizer(preprocessor = clean)
x_train_vect = vect.fit_transform(x_train)
x_test_vect = vect.transform(x_test)
```
