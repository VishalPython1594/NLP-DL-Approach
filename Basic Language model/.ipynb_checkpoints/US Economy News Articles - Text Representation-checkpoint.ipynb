```python
# removing the data where we don't want relevance ="Not Sure"
print(df.shape)
df = df[df.relevance !='not sure']
print(df.shape)
```

```python
# convert label to a numerical variable
df['relevance'] = df.relevance.map({'yes':1, 'no':0})
# relevant is 1 and non-relevant is 0

# let us take only the two columns we need
df = df[['text', 'relevance']]
df.shape
```

```python
df.head()
```

# Step 3 - Segregate inputs (x) and output (y)

```python
from sklearn.model_selection import train_test_split
x = df.text
y = df.relevance 
print(x.shape, y.shape)
```

# Step 4 - Split the data into Train and Test

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, random_state=1, test_size=0.25, stratify=y)
```

```python
y_train.value_counts()
```

```python
y_test.value_counts()
```

# Step - 5 : Apply Data Preparation on Training Data

```python
# preprocessing on training data
import string
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import re
lemmatizer = WordNetLemmatizer()
```

```python
def clean(doc):
    doc = doc.replace("</br>"," ")
    doc = "".join([char for char in doc if char not in string.punctuation and not char.isdigit()])
    tokens = nltk.word_tokenize(doc)
    lemmatized_tokens = [lemmatizer.lemmatize(token) for token in tokens]
    stop_words = set(stopwords.words('english'))
    filtered_tokens = [word for word in lemmatized_tokens if word.lower() not in stop_words]
    return " ".join(filtered_tokens)
```

```python
# import feature extraction methods from sklearn
# Bag of Words
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
# convert to a vectoriezer
vect = CountVectorizer(preprocessor=clean)
# use it to extract features from training data
%time x_train_dtm = vect.fit_transform(x_train)
print()
print(f"shape of x_train_transformed (# of docs, # of unique vocabulary): {x_train_dtm.shape}")
print(f"Type of x_train_transformed (i.e. compressed Sparse Row (CSR) format): {type(x_train_dtm)}")
print(f"Vocabulary learned : {vect.get_feature_names_out()[500:510]}")
```

```python
print(f"Vocabulary learned : {vect.get_feature_names_out()[5000:5020]}")
```

```python
def tokenizer(doc):
    return nltk.word_tokenize(doc)

from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
bow_vect = CountVectorizer(token_pattern=None,tokenizer=tokenizer, ngram_range=(1, 1),preprocessor=clean,stop_words=None)
%time x_train_transformed = bow_vect.fit_transform(x_train)
print()
print(f"shape of x_train_transformed (# of docs, # of unique vocabulary): {x_train_transformed.shape}")
print(f"Type of x_train_transformed (i.e. compressed Sparse Row (CSR) format): {type(x_train_transformed)}")
print(f"Vocabulary learned : {bow_vect.get_feature_names_out()[500:520]}")
```

```python
x_test_dtm = bow_vect.transform(x_test)
print(x_test_dtm.shape)
```

```python
x_test_dtm1 = vect.transform(x_test)
print(x_test_dtm1.shape)
```

# Building a model

```python
from sklearn.naive_bayes import MultinomialNB
nb = MultinomialNB()
%time nb.fit(x_train_dtm, y_train)
```

```python
# evaluate the train data
from sklearn import metrics
y_train_pred = nb.predict(x_train_dtm)
metrics.confusion_matrix(y_train, y_train_pred)
```

```python
metrics.accuracy_score(y_train, y_train_pred)
```

```python
print(metrics.classification_report(y_train, y_train_pred))
```

```python
cm = metrics.confusion_matrix(y_train, y_train_pred)
actual = np.sum(cm, axis=1).reshape(-1,1)
cmn = np.round(cm/actual, 2)
sns.heatmap(cmn, annot=True, fmt='.2f', xticklabels=nb.classes_, yticklabels=nb.classes_)
plt.ylabel("Actual")
plt.xlabel("Predicted")
```

```python
# Evaluate on test data
y_test_pred = nb.predict(x_test_dtm1)
metrics.confusion_matrix(y_test, y_test_pred)
```

```python
metrics.accuracy_score(y_test, y_test_pred)
```

```python
print(metrics.classification_report(y_test, y_test_pred))
```

```python
cm = metrics.confusion_matrix(y_test, y_test_pred)
actual = np.sum(cm, axis=1).reshape(-1,1)
cmn = np.round(cm/actual, 2)
sns.heatmap(cmn, annot=True, fmt='.2f', xticklabels=nb.classes_, yticklabels=nb.classes_)
plt.ylabel("Actual")
plt.xlabel("Predicted")
```

```python
print(metrics.classification_report(y_train, y_train_pred))
print()
print(metrics.classification_report(y_test, y_test_pred))
```
